# ðŸ’§ LFM2-1.2B-RAG Inference with transformers

This notebook allows you to easily run LFM2-1.2B-RAG with Hugging Face's [transformers](https://github.com/huggingface/transformers) library.

You can run it on GPU or CPU by switching the runtime (`Runtime` â†’ `Change runtime type`).

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load model and tokenizer
model_id = "LiquidAI/LFM2-1.2B-RAG"
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    dtype="bfloat16",
#    attn_implementation="flash_attention_2" <- uncomment on compatible GPU
)
tokenizer = AutoTokenizer.from_pretrained(model_id)

# Create message
user_prompt = """Use the following context to answer questions:

Desert Bloom Technologies: Pioneering Biodiversity Corridors in the UAE's Arid Research Zones â€“ A White Paper

Executive Summary:

The United Arab Emirates, while renowned for its rapid development, faces significant environmental challenges stemming from aridification, habitat fragmentation, and biodiversity loss. Desert Bloom Technologies (DBT) is a pioneering innovator focused on restoring and connecting fragmented desert ecosystems through targeted microhabitat creation and the establishment of functional biodiversity corridors within designated research zones. This paper details DBTâ€™s approach, leveraging advanced soil augmentation techniques, native species propagation, and AI-driven environmental monitoring to facilitate ecological recovery and enhance resilience in the face of climate change.

The Challenge: Fragmented Ecosystems & Biodiversity Loss

The UAEâ€™s desert landscapes, though seemingly homogenous, support a surprising diversity of specialized flora and fauna. However, urbanization, infrastructure development, and historical land use practices have created significant habitat fragmentation, isolating populations and hindering gene flow. This isolation leads to reduced genetic diversity, increased vulnerability to disease, and ultimately, species decline. Traditional restoration efforts often struggle due to harsh environmental conditions â€“ intense heat, limited water availability, and saline soils.

According to the document, what specific challenges hinder traditional restoration efforts in the UAE's desert landscapes, and how does Desert Bloom Technologies aim to address these limitations?"""
message = [
  {"role": "system", "content": "Always respond in English, regardless of the input language used by the user."}, # System prompt is optional
  {"role": "user", "content": user_prompt}
]

# Generate answer
input_ids = tokenizer.apply_chat_template(
    message,
    add_generation_prompt=True,
    return_tensors="pt",
    tokenize=True,
).to(model.device)

output = model.generate(
    input_ids,
    do_sample=False,
    max_new_tokens=512,
)

print(tokenizer.decode(output[0], skip_special_tokens=False))
```
